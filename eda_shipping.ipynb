# Shipping Data — Exploratory Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns

%matplotlib inline
sns.set_theme(style="whitegrid")
```

## 1. Load Data

```python
df = pd.read_excel("shipping_data.xlsx", parse_dates=["Ship Date", "Delivery Date"])
df.head(10)
```

## 2. Basic Info

```python
print("Shape:", df.shape)
print("\nColumn types:")
print(df.dtypes)
```

```python
print("Missing values:")
df.isnull().sum()
```

## 3. Summary Statistics

```python
df[["Transit Days", "Weight (lbs)", "Shipping Cost ($)"]].describe().round(2)
```

## 4. Categorical Counts

```python
for col in ["Carrier", "Status", "Category", "On Time"]:
    print(f"\n{col}:\n{df[col].value_counts()}\n" + "-"*30)
```

## 5. On-Time Rate & Cost by Carrier

```python
on_time_rate = (
    df.groupby("Carrier")["On Time"]
    .apply(lambda x: (x == "Yes").mean() * 100)
    .reset_index()
    .rename(columns={"On Time": "On Time Rate (%)"})
    .sort_values("On Time Rate (%)", ascending=False)
)

avg_cost = df.groupby("Carrier")["Shipping Cost ($)"].mean().sort_values(ascending=False).round(2)
avg_transit = df.groupby("Carrier")["Transit Days"].mean().sort_values().round(2)

print("On-Time Rate by Carrier:")
print(on_time_rate.to_string(index=False))
print("\nAvg Shipping Cost by Carrier:")
print(avg_cost)
print("\nAvg Transit Days by Carrier:")
print(avg_transit)
```

## 6. Plots

### Shipments by Carrier

```python
carrier_counts = df["Carrier"].value_counts()
plt.figure(figsize=(8, 4))
plt.bar(carrier_counts.index, carrier_counts.values, color=sns.color_palette("tab10"))
plt.title("Shipments by Carrier")
plt.xlabel("Carrier")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

### Shipment Status Distribution

```python
status_counts = df["Status"].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(status_counts.values, labels=status_counts.index, autopct="%1.1f%%", startangle=140)
plt.title("Shipment Status Distribution")
plt.tight_layout()
plt.show()
```

### On-Time Rate by Carrier

```python
plt.figure(figsize=(8, 4))
plt.barh(on_time_rate["Carrier"], on_time_rate["On Time Rate (%)"], color="steelblue")
plt.axvline(75, color="red", linestyle="--", label="Overall avg")
plt.title("On-Time Rate by Carrier (%)")
plt.xlabel("On-Time Rate (%)")
plt.legend()
plt.tight_layout()
plt.show()
```

### Shipping Cost Distribution

```python
plt.figure(figsize=(8, 4))
plt.hist(df["Shipping Cost ($)"], bins=30, color="coral", edgecolor="white")
plt.title("Shipping Cost Distribution")
plt.xlabel("Cost ($)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

### Weight vs Shipping Cost

```python
plt.figure(figsize=(8, 5))
plt.scatter(df["Weight (lbs)"], df["Shipping Cost ($)"], alpha=0.4, color="mediumseagreen")
plt.title("Weight vs Shipping Cost")
plt.xlabel("Weight (lbs)")
plt.ylabel("Shipping Cost ($)")
plt.tight_layout()
plt.show()
```

### Transit Days Distribution

```python
plt.figure(figsize=(8, 4))
plt.hist(df["Transit Days"], bins=14, color="mediumpurple", edgecolor="white", rwidth=0.8)
plt.title("Transit Days Distribution")
plt.xlabel("Transit Days")
plt.ylabel("Frequency")
plt.gca().xaxis.set_major_locator(mticker.MultipleLocator(1))
plt.tight_layout()
plt.show()
```

### Avg Shipping Cost by Category

```python
avg_cost_cat = df.groupby("Category")["Shipping Cost ($)"].mean().sort_values(ascending=False)
plt.figure(figsize=(8, 4))
plt.bar(avg_cost_cat.index, avg_cost_cat.values, color=sns.color_palette("pastel"))
plt.title("Avg Shipping Cost by Category")
plt.xlabel("Category")
plt.ylabel("Avg Cost ($)")
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```

### Shipments by Origin City

```python
origin_counts = df["Origin"].value_counts()
plt.figure(figsize=(8, 4))
plt.barh(origin_counts.index, origin_counts.values, color="sandybrown")
plt.title("Shipments by Origin City")
plt.xlabel("Count")
plt.tight_layout()
plt.show()
```

### Avg Shipping Cost by Carrier

```python
plt.figure(figsize=(8, 4))
plt.bar(avg_cost.index, avg_cost.values, color=sns.color_palette("muted"))
plt.title("Avg Shipping Cost by Carrier")
plt.xlabel("Carrier")
plt.ylabel("Avg Cost ($)")
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```
